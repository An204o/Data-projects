```python
import torch
import torch.nn as nn
import torch.nn.functional as F

model = nn.Sequential(
    nn.Linear(8, 4),
    nn.ReLU(),  # Corrected here
    nn.Linear(4, 5),
    nn.Softmax(dim=1)
)
```

```python
for name, param in model.named_parameters():
    if name == '0.weight' and name == '0.bias':
        param.requires_grad = False
```

```python
import pandas as pd

df = pd.read_csv('animals.csv')
df.drop('animal_name', axis = 1)
```

```python
X = df.drop('type', axis=1).drop('animal_name', axis=1)
y = df['type']
```

```python
from torch.utils.data import Dataset, DataLoader, TensorDataset

tensor = TensorDataset(torch.tensor(X.values).float(), torch.tensor(y.values))
```

```python
dataloader = DataLoader(tensor, batch_size=10, shuffle=True)
```

```python
num_epochs = 10
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
for epoch in range(num_epochs):
    for data, labels in dataloader:
      optimizer.zero_grad()
      outputs = model(data)
      loss = F.cross_entropy(outputs, labels)
      loss.backward()
      optimizer.step()
```
